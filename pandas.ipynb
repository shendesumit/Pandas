```python
def map_num_name(x):
    '''Works only for these numbers 
    ['four', 'six', 'five', 'three', 'twelve', 'two', 'eight']'''
    if x == 'four':
        return 4
    elif x == 'six':
        return 6
    elif x == 'five':
        return 5
    elif x == 'three':
        return 3
    elif x == 'twelve':
        return 12
    elif x == 'two':
        return 2
    elif x == 'eight':
        return 8
    
    
```

```python
map_num_name('four')
map_num_name('five')
map_num_name('six')
```

### using apply

```python
car_df.num_of_cylinders.apply(map_num_name)
```

```python
car_df.num_of_cylinders.apply(map_num_name)
```

### other operators

```python
k = []
for i in car_df.num_of_cylinders:
    k.append(map_num_name(i))
```
